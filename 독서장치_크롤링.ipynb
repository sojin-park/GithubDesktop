```python
from bs4 import BeautifulSoup
import urllib.request as req
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import requests
import time
import urllib.request
import pandas as pd
import os
import numpy as np
```

```python
driver = webdriver.Chrome() 
driver.get('https://nld.go.kr/home/libraryPossession.do?menu=menu_03_03&low=N')
time.sleep(3)
```

```python
'''
 - 동적페이지가 아닐경우 html으로 표 크롤링하기!!-
크롤링 참고-https://wikidocs.net/238534
ed!# "지역을 선택해 주세요" 요소 찾기
xpath = '//*[@id="__next"]/div[2]/div/div/div[2]/div[1]/div[1]'
element = driver.find_element(By.XPATH, xpath)

# 요소의 위치로 스크롤 조정
driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", element)'''
```

```python
data =[]
page=113

for i in range(page):
    
    # 표 안에 있는 항목 저장
    for td in range(1,11):
        df=[]
        td_list = ('//*[@id="list"]/tr[{}]'.format(td))
        for i in range(1,6):
            element = (driver.find_element(By.XPATH, td_list+'/td[{}]'.format(i))).text #text 는 문자로변환
            df.append(element)
        data.append(df)#list 안 list 넣어야함
            
    #다음 페이지로 넘김
    try:
        pages = f'//*[@id="paging"]/a[i])' #page 항목
        sample = driver.find_element(By.CSS_SELECTOR,'a.next') # 다음버튼만 있을 때 
        sample.click()
    except NoSuchElementException as e:
        print('마지막 페이지입니다.')
        break

data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.columns = ["번호", '도서관명', '전화번호', '보조공학기기 분류', '보조공학기기 명']
df
```

```python
new_column = []
for item in df['보조공학기기 분류']:
    if '이동' in item or '마우스' in item or '키보드' in item or '모니터' in item or '팔받침' in item or '휠체어' in item or '테이블' in item or '페이지' in item or '홀더' in item or '스위치' in item:
        new_column.append('지체·뇌병변장애') 
    elif '확대' in item or '낭독' in item or '플레이어' in item or '인쇄물음' in item or '점자' in item or '바코드' in item:
        new_column.append('시각장애')  
    elif  '증폭' in item or '의사소통' in item or '보청기' in item or '신호' in item or '수화' in item or '헤드셋' in item:
        new_column.append('청각·언어장애')
    else:
        new_column.append('확인 필요')
new_column
df['보조기기'] = new_column
```

```python
result =df[df['보조기기']== '확인 필요']
print(result)
```
